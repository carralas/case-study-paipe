```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/train_houses 1(in).csv')
```

```python
pd.options.display.max_columns = None
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.countplot(x=df.Municipality)

plt.xticks(rotation=90);
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(x=df.Municipality, y=df.TradePrice)

plt.xticks(rotation=90);
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(x=df.Municipality, y=df.TradePrice, showfliers=False)

plt.xticks(rotation=90);
```
